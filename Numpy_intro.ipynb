```python
import numpy as np
```

```python
names=np.array(["Alice","Charlie","Eugene","Hannah","Oscar","Silatan","Bobby","Richey","Mitchell","William"])
scores=np.array([22,67,56,27,60,78,10,44,99,100])
```

```python
for name,score in zip(names,scores):
    print(name,"-",score)
```

```python
indices=np.argsort(scores)
sortednames=names[indices]
sortedscores=scores[indices]
```

```python
for name,score in zip(sortednames,sortedscores):
    print(name,"-",score)
```

# **Findings**


*   **William** is the **best** student in the class with **100/100**
*   **Bobby** got the **least** and needs to improve, he has **10/100**
